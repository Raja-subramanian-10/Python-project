HYPOTHESIS TESTING:
-
           
           Hypothesis testing is a statistical method used to make decisions or inferences about a population based on sample data. 
In this we used three types of hypothesis testing , they are 
            
            * T-testing - used to compare the average of the numerical values between two groups
            
            * chi-square test - used to test association between two catergorical variables 
            
            * ANOVA - Used to compare the average of a numnerical variable across three or more groups  

```python
# --- Hypothesis Testing ---

# 1. T-test: Compare average price between "Entire home/apt" and "Private room"
entire_home_prices = NYK_clean[NYK_clean['room_type'] == 'Entire home/apt']['price']
private_room_prices = NYK_clean[NYK_clean['room_type'] == 'Private room']['price']

t_stat, p_val_ttest = stats.ttest_ind(entire_home_prices, private_room_prices, equal_var=False)
print(f"T-test (Entire home vs Private room): t-stat = {t_stat:.2f}, p-value = {p_val_ttest:.2e}")
if p_val_ttest < 0.05:
    print("✅ Significant difference in average prices between room types.\n")
else:
    print("❌ No significant difference in average prices between room types.\n")
```

INSIGHTS:
-


Statistical Significance: 
            
            The p-value is far below any conventional threshold (e.g., 0.05), so we  can confidently reject the null hypothesis.
    → There is an difference in average prices between Entire home/apt and Private room listings.

Direction of Difference: 
            
            Since the t-stat is positive and you're comparing Entire home vs Private room.
    → Entire homes/apartments are significantly more expensive than private rooms.

```python
# 2. Chi-Square test: Relationship between neighbourhood_group and room_type
contingency_table = pd.crosstab(NYK_clean['neighbourhood_group'], NYK_clean['room_type'])
chi2, p_val_chi2, dof, expected = stats.chi2_contingency(contingency_table)
print(f"Chi-Square Test: chi2 = {chi2:.2f}, df = {dof}, p-value = {p_val_chi2:.2e}")
if p_val_chi2 < 0.05:
    print("✅ Neighbourhood group and room type are dependent.\n")
else:
    print("❌ No significant association between neighbourhood group and room type.\n")
```

INSIGHTS:
-

Statistical Significance: 

    The p-value is astronomically small, so we reject the null hypothesis of independence.

Interpretation: 

    The distribution of room types (Entire home/apt, Private room, Shared room) varies significantly across neighbourhood groups (Manhattan, Brooklyn, Queens, Bronx, Staten Island).

```python
# 3. ANOVA: Difference in price among neighbourhood_groups
anova_groups = [NYK_clean[NYK_clean['neighbourhood_group'] == group]['price'] 
                for group in NYK_clean['neighbourhood_group'].unique()]
f_stat, p_val_anova = stats.f_oneway(*anova_groups)
print(f"ANOVA: F-stat = {f_stat:.2f}, p-value = {p_val_anova:.2e}")
if p_val_anova < 0.05:
    print("✅ Significant differences in prices across neighbourhood groups.\n")
else:
    print("❌ No significant differences in prices across neighbourhood groups.\n")
```

INSIGHTS
-

F-statistic = 22.74 

This tells us that the variation in prices between neighbourhoods is much larger than the variation within each neighbourhood.

p-value = 1.02e-18 

That’s an extremely small number—way below the usual threshold of 0.05. It means the result is statistically significant.
